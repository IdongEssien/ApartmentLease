#### 4.0 The stationarity Test

```python
# Importing the needed library for the test

from statsmodels.tsa.stattools import adfuller
```

```python
# Defining the function

timeseries = data_sub['Value'].values
```

```python
# Performing the Dicker-Fuller test

print ('Dickney-Fuller Stationarity test:')
test = adfuller(timeseries, autolag='AIC')
result = pd.Series(test[0:4], index=['Test Statistic', 'p-value', 'Number of Lags Used', 'Number of Observations Used'])
for key,value in test[4].items():
    result['Critical Value (%s)'%key] = value
    print (result)
```

```python
# Using another methodology (ADF) to double check the efficacy of the Dickey-Fuller Test

result = adfuller(timeseries)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
    
if result[0] < result[4]['5%']:
    print ('Reject null hypothesis - Time series is stationary')
else:
    print ('Cannot reject null hypothesis - Time series is non-stationary')
```

- Both tests produced the same result. Unfortunately, the test statistic value of -1.161616 is higher than the 5% critical value (significance level) which is -2.972. It means we cannot reject the null hypothesis as the time series data is non-stationary. This is quite evident from the upward trajectory of the plotted line chart from the time-series data.
- Techniques would have to be carried out to stationarize the data

##### 5.0 Stationarizing the data
- 5.1 Differencing

```python
# Creating a new subset as 'data_diff' for the purpose of differencing 

data_diff = data_sub - data_sub.shift(1)
```

```python
# Eliminating possible missing values that may arise from the differencing implementation

data_diff.dropna(inplace = True)
```

```python
data_diff.head()
```

```python
data_diff.columns
```

```python
# Plotting the new line graph after the differencing

plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff)
```

- The resultant line chart is visibly changed from the original version and already portrays elements of stationarity. But it will still be subjected to testing

```python
# Creating a new timeseries(as timeseries2) for the data_diff subset to prevent overwriting the timeseries of data_sub

timeseries2 = data_diff['Value'].values
```

```python
# Carrying out the ADF test

result = adfuller(timeseries2)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
    
if result[0] < result[4]['5%']:
    print ('Reject null hypothesis - Time series is stationary')
else:
    print ('Cannot reject null hypothesis - Time series is non-stationary')
```

- Voila!!!!! The new test statistic is lower than the critical value (-3.689 < -2.972). So we can reject the null hypothesis and confirm that the diferencing method has made the data stationary 

##### 6.0 Autocorrelation

```python
# Importing data for plotting the autocorrelation graph

from statsmodels.graphics.tsaplots import plot_acf
```

```python
# Plotting the graph of the details from the data_diff subset

plot_acf(data_diff)
plt.show()
```

- All the vertical lines fall within the confidence interval represented by the blue area. An indication that the staionarity after implementing the differencing is strong.
