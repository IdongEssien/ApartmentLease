## 1.0 Heatmap Creation

```python
# Selecting columns to create a subset (as df_NY) for the heatmap and other plots

df_NY = df_Airbnb[['neighbourhood_group', 'room_type', 'construction_year', 'price', 'service_fee',
       'minimum_nights', 'rating', 'host_listings_count']]
```

```python
df_NY.head()
```

```python
df_NY.shape
```

```python
cor = df_NY.corr()
cor
```

```python
sns.heatmap(data = cor)
```

##### Observation from the heatmap: 
- From the heatmap (and indicator),it is instantly apparent that there is no relationship among the generality of the variables
- EXCEPT between 'price' & 'service_fee'where the relationship is very strong with a coefficient of 1.
- 'Minimum_nights' and 'host_listings_count' showed a glimmer of hope, but their relationship is still very week.

## 2.0 Scatter Plot Creation

```python
scatplot = sns.scatterplot(x = 'price', y = 'service_fee', data = df_NY)
```

##### Observation from the Scatterplot:
- From the plottings above, there is no gain saying the resultant coefficient is 1.

## 4.0 Categorical Plot creation

```python
sns.catplot(x = 'rating', y = 'minimum_nights', data = df_NY)
```

##### Result Interpretation:
- This Categorical Plot portrays a cluster of values below 50 nights, which trickles vertically upwards. Each rating still has minimum nights of over 350. The cluster is in tandem with the hypothesis that Airbnb bookings by clients range from a couple of nights to a few weeks.
