```python
# Import of needed libraries

import pandas as pd
import numpy as np
import os
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn 
from sklearn.cluster import KMeans
import pylab as pl
```

```python
# Import of dataframe as df_Airbnb

df_Airbnb = pd.read_csv(r'C:\Users\IDONG\Prepared data\6.4_Airbnb_Clean.csv', index_col = 0)
```

```python
# Display of available columns of the dataframe

df_Airbnb.columns
```

##### 1.0 Creation of the subset for the exercise (only numerical values)

```python
# Creating a subset for non-categorical variables (as df_NYC).
# Renaming of columns not required,cleaning and dataframe manipulations already carried out initially.

df_NYC = df_Airbnb[['construction_year', 'price', 'service_fee', 'minimum_nights', 'rating', 'host_listings_count']]
```

```python
# Standardizing all selected variables in the subset to eliminate data bias

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# The new standardized set will be represented as df_NYC_std

df_NYC_std = scaler.fit_transform(df_NYC)
```

```python
# Defining the range for potential clusters in the subset
num_cl = range(1, 10)

# Defining kmeans clusters in the range assigned
kmeans = [KMeans(n_clusters=i) for i in num_cl]
```

```python
# Creating the ten possible sets of clusters

score = [kmeans[i].fit(df_NYC_std).score(df_NYC_std) for i in range(len(kmeans))]
```

```python
score
```

##### 2.0 Creating the Elbow Curve

```python
# Establishing the Elbow Curve that will determine the number of clusters to be used (n_clusters=?)

pl.plot(num_cl,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

- Studying the curve, it almost appears the curve begins to flatten up just after point 3 when traced on the x-axis.
- Thus 3 was chosen as the optimal count for the number of clusters

```python
# Creating the k-means object

kmeans = KMeans(n_clusters=3)
```

```python
# Creating a new column called 'cluster' for the initial df_NYC dataframe

df_NYC['clusters'] = kmeans.fit_predict(df_NYC_std)
```

```python
# Dataframe showing the new column creation

df_NYC.head()
```

```python
# Frequency showing how values are spread among the clusters

df_NYC['clusters'].value_counts()
```

##### 3.0 Creating Scatterplot with color delineation
- 3.1 Minimum_nghts Vs. host_listings_count)

```python
# Creation of the clusters with color delineation

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df_NYC['minimum_nights'], y=df_NYC['host_listings_count'], hue=kmeans.labels_, s=100)
ax.grid(False)
plt.xlabel('minimum_nights')
plt.ylabel('host_listings_count')
plt.show()
```

##### Observation
- From the visual, the light pink cluster (code 0) is dominantly represented by plots of above 125, both on the x-axis (minimum_nights) and on the y-axis(host_listings_count).
- Below these points (of 125) also on both axes, the other two clusters of code 1 and 2 (pink and dark purple respectively) tend to juxtapose upon each other. However there are still skirmishes of light pink plots in this quadrat.

- 3.2 Construction Vs. minimum_nights

```python
# Creation of the clusters with color delineation

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df_NYC['construction_year'], y=df_NYC['minimum_nights'], hue=kmeans.labels_, s=100)
ax.grid(False)
plt.xlabel('construction_year')
plt.ylabel('minimum_nights')
plt.show()
```

- 3.3 Price Vs. minimum_nights

```python
# Creation of the clusters with color delineation

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df_NYC['price'], y=df_NYC['minimum_nights'], hue=kmeans.labels_, s=100)
ax.grid(False)
plt.xlabel('price')
plt.ylabel('minimum_nights')
plt.show()
```

##### Observation
- Now for this scatterplot, we have well defined cluster segmentation for each of the clusters. The light pink cluster maintains its appearance above 125 on the y-axis, but runs across the range of values on the x-axis. It is also scantily populated as compared to the lower end. An obesrvation similar to previous plots.
- Below 125 on the y-axis has what appears to be an even distribution of plots between the pink and dark purple clusters. Below 640 on the x-axis is the occupation of the pink cluster (code 1), while above 640 is that of the dark purple cluster.
- This is a clear cut revelation that the code 2 cluster is more expensive than its code 1 counterpart. 

##### 4.0 Descriptive Stats

```python
# Creating descriptive stats for the clusters

df_NYC.loc[df_NYC['clusters'] == 2, 'cluster'] = 'dark purple'
df_NYC.loc[df_NYC['clusters'] == 1, 'cluster'] = 'purple'
df_NYC.loc[df_NYC['clusters'] == 0, 'cluster'] = 'pink'
```

```python
df_NYC.groupby('cluster').agg({'construction_year':['mean', 'median'], 'price':['mean', 'median'], 'service_fee':
                               ['mean', 'median'], 'minimum_nights':['mean', 'median'], 'rating':['mean', 'median'], 
                              'host_listings_count':['mean', 'median']})
```

- 4.1 Insights 

In comparison with other clusters, the pink cluster segment has the highest mean rating, minimum_nights as well as host_listings_count, but not the most expensive (price/service_fee). 

From the last visual as well as from the stats table, the dark purple cluster is the most expensive of available Airbnb New York accomodation. The cluster also has the lowest mean value for minimum_nights and rating.

Lastly, the purple cluster with code 1 is the least expensive also with the least mean value of host_listings_count.
