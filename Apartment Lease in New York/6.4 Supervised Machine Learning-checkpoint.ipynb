```python
# Import libraries

import pandas as pd
import numpy as np
import os
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Import of cleaned data frame as df_Airbnb

df_Airbnb = pd.read_csv(r'C:\Users\IDONG\Prepared data\6.4_Airbnb_Clean.csv', index_col = 0)
```

```python
# Selecting variables of key interest to form a subset as df_NYC

df_NYC = df_Airbnb[['price', 'service_fee', 'minimum_nights', 'host_listings_count']]
```

## 1.0 Variables - 'price' & 'service_fee'

```python
# Creating a scatterplot of 'price' Vs 'service_fee'

scatplot = sns.scatterplot(x = 'price', y = 'service_fee', data = df_NYC)
```

##### Observation 
- 'price' and 'service_fee' have already been established to have a linear relationship from previous exercise, as the latter is a mathematical derivative of the former

```python
# Reshaping the variables

X = df_NYC['price'].values.reshape(-1,1)
y = df_NYC['service_fee'].values.reshape(-1,1)
```

```python
# Splitting the data into 2 sets: the train and the test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# Running a linear regression

regression = LinearRegression()
```

```python
# Fitting the regression to the training set

regression.fit(X_train, y_train)
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
```

```python
# Predicting y values on the test set

y_predicted = regression.predict(X_test)
```

```python
# Creating a plot with the regression line

plot_test = plt
plot_test.scatter(X_test, y_test, color='gray', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('price vs service_fee (Test set)')
plot_test.xlabel('price')
plot_test.ylabel('service_fee')
plot_test.show()
```

##### Observation
- Without having to rely on any form of interpretation, the relationship between the variables is harmonious with an R2 score of approximately 1. But we will still go further to take the measurement.

```python
# To check and confirm the model performance statistics:

rmse = mean_squared_error(y_test, y_predicted)
r2 = r2_score(y_test, y_predicted)
```

```python
print('Mean squared error:', rmse)
print('R2 score:', r2)
```

```python
# Making a comparison between predicted and actual y values

data = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_predicted.flatten()})
data.head(30)
```

##### Reflections-
- For every actual value in comparison to its predicted counterpart, the difference is not more than 0.5. This is evidence that the model has reconfirmed the exixtence of a linear relationship between the variables. The same is buttressed with results from the calculation of the Mean squared error (0.081) and R2 score (0.99)
